## FINAL CODE

```python
#now we  getting data from 2010 to 2024 dynamically.
# we noticed that the URL changes by year wise only so we will use the range function
final = pd.DataFrame()
all_numbers=[]
all_grey =[]

for j in range( 2010, 2025):
    url = 'https://www.lottodatabase.com/lotto-database/canadian-lotteries/lotto-649/draw-history/{}'.format(j)  
    webpage = requests.get(url).text
    soup = BeautifulSoup( webpage, 'lxml')
    draw_date = soup.find_all('div', class_ ='section group' )
    draw_date_r = draw_date[2:len(draw_date)-1] #slicing only the required fields
    date = []
    number1 = []
    number2 = []
    number3 = []
    number4 = []
    number5 = []
    number6 = []
    
    
    grey_b = []
    

    for i in draw_date_r :
        date.append(i.find(class_ = 'col s_3_12').text.strip())
        number1.append(i.find_all_next(class_= "white ball")[0].text)  #find_all_next gets 2d list , hence we using indexing to enter the list.
        all_numbers.append(i.find_all_next(class_= "white ball")[0].text)

        number2.append(i.find_all_next(class_= "white ball")[1].text)
        all_numbers.append(i.find_all_next(class_= "white ball")[1].text)

        number3.append(i.find_all_next(class_= "white ball")[2].text)
        all_numbers.append(i.find_all_next(class_= "white ball")[2].text)

        number4.append(i.find_all_next(class_= "white ball")[3].text)
        all_numbers.append(i.find_all_next(class_= "white ball")[3].text)

        number5.append(i.find_all_next(class_= "white ball")[4].text)
        all_numbers.append(i.find_all_next(class_= "white ball")[4].text)

        number6.append(i.find_all_next(class_= "white ball")[5].text)
        all_numbers.append(i.find_all_next(class_= "white ball")[5].text)

        grey_b.append(i.find_all_next(class_= "grey ball")[0].text)
        all_grey.append(i.find_all_next(class_= "grey ball")[0].text)

    d= {'Draw_date' : date, "number1" : number1, "number2" : number2, "number3" : number3, "number4" : number4, "number5" : number5, "number6" : number6, 'grey_ball' : grey_b  }

    df =pd.DataFrame(d) #Creates a DF for that specific year and then appends it to "final" dataframe.
    
    final = final.append(df, ignore_index = True) 
    
```
